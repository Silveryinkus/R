**PROJECT: Movie Domestic Rate % Gross**

The movie reviews website was very happy with your deliverable for the previous assignment and now they have a new request for you.

The previous consultant had created a chart for them which is illustrated on the next slide. However, the R code used to create the diagram has since been lost and cannot be recovered.Your task is to come up with the code that will re-create the same chart making it look as close as possible to the original.

A new dataset has been supplied.

```r
#recreate the visualization given to you by the 
mov <- read.csv("Section6-Homework-data.csv")
#data exploration
head(mov)
tail(mov)
summary(mov)
str(mov)
```

```r
#now let's determine what variables we will need to create
#so we need 4 columns

#load ggplot2 package
#install.packages("ggplot2")
library(ggplot2)
```

```r
#interesting insight
#it looks like most movies are released on Friday (a strategy)
#no movie was released on a monday
ggplot(data=mov, aes(x=Day.of.Week)) + geom_bar()
```

```r
#filter the dataframe for the variables that we need
filt <- (mov$Genre=="action") |(mov$Genre=="adventure") | (mov$Genre=="animation") | (mov$Genre=="comedy") | (mov$Genre=="drama")

#filter2
filt2 <-mov$Studio %in% c("Buena Vista Studios", "WB", "Fox", "Universal", "Sony", "Paramount Pictures")

mov2 <- mov[filt & filt2,]
head(mov2)
```

```r
#prep the plot's data and aes layers
p <- ggplot(data=mov2, aes(x=Genre, y=Gross...US))

#add geometrics
p +
  geom_jitter() +
  geom_boxplot(alpha=0.7)
```

```r
#add colour
p +
  geom_jitter(aes(size=Budget...mill., colour=Studio)) +
  geom_boxplot(alpha=0.7)
```

```r
#there are some dots from geom_plot() stll present, let's get rid of it
q <- p +
  geom_jitter(aes(size=Budget...mill., colour=Studio)) +
  geom_boxplot(alpha=0.7, outlier.colour=NA )

q
par(xpd=TRUE)
```

```r
#non-data ink
q <- q +  
  xlab("Genre") +
  ylab("Gross % US") +
  ggtitle("Domestic Gross % by Genre") 

q
```

```r
#theme
q <- q +  
  theme(axis.title = element_text(colour="Blue", size=10),
        axis.title.y = element_text(colour="Blue", size=10),
        axis.text.x = element_text(size=10),
        axis.text.y = element_text(size=10),
        
        legend.title = element_text(size=10),
        legend.text = element_text(size=10),
        legend.position = "right", #this moves legend outside plot
        plot.title = element_text(size=20)
        )
q        
```

```r
#Final Touch
q$labels$size <- "Budget(Million)"

q
## note: Error is due to the fact that data is run in jupyter and not R studio
```
